# Medidas de dispersion
Rango, Desviación estándar, Varianza

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('laptop_price.csv', encoding='iso-8859-1')
```

```python
# Desviación estandar
df['Price_euros'].std()
```

```python
# Rango = valor max - valor min
df['Price_euros'].max() - df['Price_euros'].min()
```

```python
# Cuartiles
median = df['Price_euros'].median()
q1 = df['Price_euros'].quantile(q=0.25)
q3 = df['Price_euros'].quantile(q=0.75)
min_val = df['Price_euros'].quantile(q=0)
max_val = df['Price_euros'].quantile(q=1.0)
print("Min=", min_val, "Q1=", q1, "Q2=", median, "Q3=", q3, "Max=", max_val)
```

```python
# Rango intercuartil
iqr = q3 - q1
iqr
```

Covarianza y correlacion

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
titanic = sns.load_dataset('titanic')
```

```python
sns.pairplot(titanic, hue='age')
```

```python
titanic.columns
```

```python
scaler = StandardScaler()
scaled = scaler.fit_transform(
    titanic[['survived', 'age', 'sibsp', 'parch', 'fare']]
)

scaled.T
```

Aqui hacemos la matriz para la covarianza

```python
covariance_matriz = np.cov(scaled.T)
covariance_matriz
```

```python
plt.figure(figsize=(10, 10))
sns.set(font_scale=1.5)
hm = sns.heatmap(covariance_matriz,
                 cbar=True,
                 annot=True,
                 square=True,
                 fmt='.2f',
                 annot_kws={'size': 12},
                 yticklabels=['survived', 'age', 'sibsp', 'parch', 'fare'],
                 xticklabels=['survived', 'age', 'sibsp', 'parch', 'fare'])
```
